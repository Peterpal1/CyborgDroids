```python
# importing pandas library
import pandas as pd
```

```python
# Remove rows with NaN values from the entire DataFrame
data_cleaned = data.dropna()

# Display the cleaned DataFrame
print(data_cleaned)
```

```python
data_cleaned.isna()
```

```python
# read the data frpm the csv
study=pd.read_csv('/content/drive/MyDrive/field_of_study.csv')
print(study)
```

```python
study.isna()
```

```python
# Assuming 'year' is the common column
merged_data = pd.merge(data, study, on='year')

# Display the merged DataFrame
print(merged_data)
#saving to csv
df = merged_data
df.to_csv('/content/drive/MyDrive/merged_data',index = True)
```

```python
# Print the first 10 rows
print(merged_data.head(10))
```

```python
# Print the last 10 rows
print(merged_data.tail(10))
```

```python
# Assuming 'merged_data' is your DataFrame
print(merged_data.describe())
```

```python
# Display basic information about the DataFrame
print(merged_data.info())
```

```python
# Plot histograms for numerical columns
import matplotlib.pyplot as plt
merged_data.hist(bins=20, figsize=(15, 10))
plt.show()
```

```python
# Display unique values in categorical columns
for column in merged_data.select_dtypes(include=['object']).columns:
    print(f"Unique values in {column}: {merged_data[column].unique()}")
```

```python
# Display the correlation matrix
print(merged_data.corr())
```

```python
# Display the count of missing values in each column
print(merged_data.isnull().sum())
```
